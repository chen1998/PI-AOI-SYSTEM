```python
from sqlalchemy import create_engine, text, inspect
from sqlalchemy.exc import SQLAlchemyError
def get_table_by_days(dbhandler, table_name, cols=None, day=None, time_key=None):
    """
    取得指定資料表在最近 N 天內的特定欄位資料。

    :param table_name: 資料表名稱
    :param cols: 欄位名稱 list 或字串；None 表示全部欄位
    :param day: 數值 → 取距離現在 N 天內資料；字串或 None → 不篩選日期
    :param time_key: 要用來篩選時間的欄位名稱（需要是 datetime/timestamp 類型）
    :return: pandas.DataFrame
    """
    try:
        
        if cols is None:
            col_sql = "*"
        elif isinstance(cols, (list, tuple, set)):
            safe_cols = [f"`{c}`" for c in cols]
            col_sql = ", ".join(safe_cols)
        else:
           
            col_sql = str(cols)

        
        base_sql = f"SELECT {col_sql} FROM `{table_name}`"
        params = {}

        
        if isinstance(day, (int, float)) and time_key:
            cutoff = datetime.now() - timedelta(days=float(day))
            base_sql += f" WHERE `{time_key}` >= :cutoff"
            params["cutoff"] = cutoff
            print(
                f"[get_table_by_days] 從 '{table_name}' 取得最近 {day} 天內資料（time_key='{time_key}'）."
            )
        else:
            print(
                f"[get_table_by_days] 從 '{table_name}' 取得全資料（不篩選日期）."
            )

        with dbhandler.engine.connect() as conn:
            df = pd.read_sql(text(base_sql), conn, params=params)
        print(f"[get_table_by_days] 讀取完成：{len(df)} rows.")
        return df

    except SQLAlchemyError as e:
        print(f"[get_table_by_days] 讀取 '{table_name}' 發生錯誤: {e}")
        return pd.DataFrame()
```

```python
"""
後端
### 計算aoi tool使用率
by aoi資料表讀讀取 f'aoi_summary_{aoi}'
過濾: 依照 aoi by recipe_id
分群: by aoi(根據資料表名稱) -> day-> hour 
計數: by每小時計算glass片數(同glass_id有重複也要計算)


前端
CHART + FILTER +  + TABLE + 編輯區域
chart:分別顯示每個aoi機台每天的總使用率
table:觸發特定aoi的特定天柱狀後顯示對應資料
aoi|0|1|2|...24小時
---

"""

ori_fix_columns = [ 'run_day', 'scantime', 'glass_id', 'recipe_id' ] # 資料表保留欄位
capa_val_cfg = {
    'aoi100': 168,
    'aoi200': 238,             #recipe_id第一個字元符合者
    'aoi300': 203 # recipe_id == 'CELL-ITO'或包含後三者隨意一字串
}
```

```python
from datetime import timedelta
import numpy as np
import pandas as pd

group_keys = ['aoi', 'run_day', 'hour', 'pi_type']  # 如果之後還要用可以保留
chart_day = 30

# 1) 每天總使用率（給日 chart 用）
capa = pd.DataFrame()

# 2) 每小時使用率（含 pi_type & ALL）
hourly_capa_all = pd.DataFrame()

# 每台 AOI 的每日 target 片數
capa_val_cfg = {
    'aoi100': 168,
    'aoi200': 238,
    'aoi300': 203,
}

# 建一個「chart_day 天 × 24 小時」的空框（給 clean.empty 用）
def make_empty_frames_for_aoi(aoi: str, chart_day: int):
    target_count = capa_val_cfg.get(aoi, np.nan)

    today = pd.Timestamp.today().normalize()
    start = today - timedelta(days=chart_day - 1)
    days = pd.date_range(start=start, end=today, freq="D")

    # (1) 每日 capa：全 0
    day_df = pd.DataFrame({
        "aoi": aoi,
        "run_day": days.date,
        "total_glass": 0,
        "target_count": target_count,
        "real_day_capa": 0.0,
    })

    # (2) 每小時 capa：只做 ALL（一台機台、無區分 pi_type）
    records = []
    for d in days.date:
        cumu = 0
        for h in range(24):
            records.append({
                "aoi": aoi,
                "run_day": d,
                "hour_int": h,
                "pi_type": "ALL",
                "hour": 0,              # 該小時片數
                "cumu": 0,              # 累計片數
                "real_hour_capa": 0.0,  # 該小時使用率
                "real_cumu_capa": 0.0,  # 累計使用率
            })
    hour_df = pd.DataFrame.from_records(records)
    return day_df, hour_df

ori_fix_columns = [ 'run_day', 'scantime', 'glass_id', 'recipe_id' ] # 資料表保留欄位
capa_val_cfg = {
    'aoi100': 168,
    'aoi200': 238,             #recipe_id第一個字元符合者
    'aoi300': 203 # recipe_id == 'CELL-ITO'或包含後三者隨意一字串
}

for aoi in cfg.aoi_names:
    aoi_summary_tbn = f'aoi_summary_{aoi}' if aoi !='aoi300' else  'aoi_summary_aoi300_capa'
    tb = get_table_by_days(dbhandler, aoi_summary_tbn, ori_fix_columns, chart_day, 'scantime')
    print(aoi_summary_tbn, len(tb))

    # 1) 時間欄位處理
    tb["scantime"] = pd.to_datetime(tb["scantime"], errors="coerce")
    tb["hour"] = tb["scantime"].dt.strftime("%Y-%m-%d %H")
    if 'run_day' not in tb.columns:
        tb["run_day"] = tb["scantime"].dt.date

    # 2) AOI 分類 + pi_type
    if aoi in ['aoi100', 'aoi200']:
        charts = ['2', '3', '4', '5']
        tb = tb[tb['recipe_id'].astype(str).str[0].isin(charts)]
        tb['pi_type'] = tb['recipe_id'].astype(str).str[0].map({
            '2': 'API',
            '4': 'API',
            '3': 'BPI',
            '5': 'BPI'
        }).fillna('')

    elif aoi == 'aoi300':
        charts = ['CELL-ITO', 'API', 'BPI', 'ITO']
        tb = tb[
            tb['recipe_id'].astype(str).apply(
                lambda x: x == charts[0] or any(c in x for c in charts[1:])
            )
        ]
        conditions = [
            tb['recipe_id'].str.contains('API', na=False),
            tb['recipe_id'].str.contains('BPI', na=False),
            tb['recipe_id'].str.contains('CELL-ITO', na=False),
            tb['recipe_id'].str.contains('ITO', na=False),
        ]
        choices = ['API', 'BPI', 'ITO', 'ITO']
        tb['pi_type'] = np.select(conditions, choices, default=None)

    # Debug：看 pi_type 分布（要的話可以留著）
    if not tb.empty:
        print({r: len(rs) for r, rs in tb.groupby(['pi_type'])})
        print(tb.iloc[0, :].to_dict())

    clean = tb.copy()
    clean = clean.dropna(subset=["scantime"])

    # ====== 如果 clean 完全沒資料 → 建 chart_day 份空表格 ======
    if clean.empty:
        print(f"[{aoi}] 此區間無資料，建立 chart_day 空框。")
        day_df, hour_df = make_empty_frames_for_aoi(aoi, chart_day)

        # 串進全域 capa / hourly_capa_all
        capa = day_df if capa.empty else pd.concat([capa, day_df], ignore_index=True)
        hourly_capa_all = hour_df if hourly_capa_all.empty else pd.concat([hourly_capa_all, hour_df], ignore_index=True)
        # 換下一台 AOI
        continue

    # ====== A. 有資料：先算每小時 glass_id 片數 ======
    cnt = (
        clean
        .groupby(["run_day", "hour", "pi_type"], dropna=False)["glass_id"]
        .size()
        .rename("glass_count")
        .reset_index()
    )

    # 建完整 hour × pi_type
    min_day = clean["run_day"].min()
    max_day = clean["run_day"].max()
    all_hours = pd.date_range(
        start=pd.to_datetime(min_day),
        end=pd.to_datetime(max_day) + timedelta(days=1) - timedelta(hours=1),
        freq="H"
    )

    pi_types = sorted(clean["pi_type"].dropna().unique())
    if not pi_types:
        pi_types = [""]

    full_idx = pd.MultiIndex.from_product(
        [all_hours, pi_types],
        names=["hour_ts", "pi_type"]
    )

    full_frame = pd.DataFrame(index=full_idx).reset_index()
    full_frame["run_day"] = full_frame["hour_ts"].dt.date
    full_frame["hour"] = full_frame["hour_ts"].dt.strftime("%Y-%m-%d %H")
    full_frame.drop(columns=["hour_ts"], inplace=True)

    hourly_usage = (
        full_frame
        .merge(cnt, on=["run_day", "hour", "pi_type"], how="left")
        .sort_values(["run_day", "hour", "pi_type"])
        .reset_index(drop=True)
    )
    hourly_usage["aoi"] = aoi

    # 取得「小時」的整數 0~23
    hourly_usage["hour_int"] = pd.to_datetime(hourly_usage["hour"]).dt.hour

    # ====== B. 做「累積片數」 + 每小時/累積使用率 ======

    # NaN 視為 0 來做累積
    hourly_usage["glass_count_f"] = hourly_usage["glass_count"].fillna(0)

    # 1) 依 pi_type 各自累積（每個 day 從 0 點累計到 23 點）
    hourly_usage = hourly_usage.sort_values(
        ["aoi", "run_day", "pi_type", "hour_int"]
    )
    hourly_usage["glass_pi"] = (
        hourly_usage
        .groupby(["aoi", "run_day", "pi_type"], dropna=False)["glass_count_f"]
        .cumsum()
    )

    # 2) 不分 pi_type 的每小時 & 累積（ALL）
    tmp_all = (
        hourly_usage
        .groupby(["aoi", "run_day", "hour_int"], as_index=False)["glass_count_f"]
        .sum()
        .rename(columns={"glass_count_f": "glass_count_all"})
    )
    tmp_all = tmp_all.sort_values(["aoi", "run_day", "hour_int"])
    tmp_all["glass_all"] = (
        tmp_all
        .groupby(["aoi", "run_day"], dropna=False)["glass_count_all"]
        .cumsum()
    )

    target_count = capa_val_cfg.get(aoi, np.nan)

    # pi_type 切分：每小時 & 累積使用率
    hourly_usage["target_count"] = target_count
    hourly_usage["real_hour_capa"] = (hourly_usage["glass_count_f"] / target_count).round(2)
    hourly_usage["real_cumu_capa"] = (hourly_usage["glass_pi"] / target_count).round(2)

    # ALL
    tmp_all["target_count"] = target_count
    tmp_all["real_hour_capa"] = (tmp_all["glass_count_all"] / target_count).round(2)
    tmp_all["real_cumu_capa"] = (tmp_all["glass_all"] / target_count).round(2)

    # ====== C. 每天總使用率 real_day_capa（用 ALL 最後一個 hour_int 的 glass_all） ======
    day_last = (
        tmp_all
        .sort_values(["aoi", "run_day", "hour_int"])
        .groupby(["aoi", "run_day"], as_index=False)
        .tail(1)
    )
    day_total = day_last[["aoi", "run_day", "glass_all", "target_count"]].copy()
    day_total.rename(columns={"glass_all": "total_glass"}, inplace=True)
    day_total["real_day_capa"] = (day_total["total_glass"] / day_total["target_count"]).round(2)

    capa = day_total if capa.empty else pd.concat([capa, day_total], ignore_index=True)

    # ====== D. 整理成「每小時使用率 table」：pi_type & ALL ======

    # pi_type（API/BPI/ITO...）
    hourly_pi_out = hourly_usage[[
        "aoi", "run_day", "hour_int", "pi_type",
        "glass_count_f", "glass_pi",
        "real_hour_capa", "real_cumu_capa"
    ]].copy()
    hourly_pi_out.rename(columns={
        "glass_count_f": "hour",   # 該小時片數
        "glass_pi": "cumu"         # 累計片數
    }, inplace=True)

    # ALL（不分 pi_type）
    hourly_all_out = tmp_all[[
        "aoi", "run_day", "hour_int",
        "glass_count_all", "glass_all",
        "real_hour_capa", "real_cumu_capa"
    ]].copy()
    hourly_all_out["pi_type"] = "ALL"
    hourly_all_out.rename(columns={
        "glass_count_all": "hour",
        "glass_all": "cumu"
    }, inplace=True)

    hourly_capa = pd.concat([hourly_pi_out, hourly_all_out], ignore_index=True)
    hourly_capa_all = hourly_capa if hourly_capa_all.empty else pd.concat([hourly_capa_all, hourly_capa], ignore_index=True)
    #break
# ==== 迴圈結束後（給 chart / table 用）====
print("每日 real_day_capa：")
print(capa.head())

print("每小時 capa (含 ALL)：")
print(hourly_capa_all.head())
```
